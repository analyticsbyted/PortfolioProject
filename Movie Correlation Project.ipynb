```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import re

%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\ALI\Documents\Python Scripts\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
   pct_missing =  np.mean(df[col].isnull())
   print(col, '-', pct_missing,'%')
```

```python
# Removing rows to missing values
df.dropna(inplace=True)
```

```python
# Data types for our columns

df.dtypes
```

```python
# Convert data type of columns 'budget' and 'gross' from float to integer.

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Lets create the correct Year column in the original data frame

df['yearcorrect'] = [re.findall(r'\d{4}', str(x)) for x in df['released']] 
```

```python
df['yearcorrect'] = df['yearcorrect'].str[0]
```

```python
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', 500)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
# Budget high correlation
# company high correlation
```

```python
# Lets build a scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earning')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
```

```python
df.corr() #pearson, kendall, spearman
```

```python
# High correlation between budget and gross.
# I was right

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Lets numerize categorical columns

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
